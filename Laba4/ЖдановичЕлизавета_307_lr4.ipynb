### 3. Построение Байесовсой сети

Данный метод представляет собой конструктивное построение структуры дискретной байесовской сети путем явного задания направленных ребер между узлами. В этом подходе структура сети создается декларативно - мы напрямую указываем список кортежей (родительский узел, дочерний узел), что определяет направленные зависимости между переменными.

````python
model = BayesianNetwork(best_structure.edges())
model.fit(train_df, estimator=MaximumLikelihoodEstimator)
````

````python
from pgmpy.models import BayesianNetwork

expert_edges = [
    ('hair', 'class'),
    ('feathers', 'class'),
    ('eggs', 'class'),
    ('milk', 'class'),
    ('airborne', 'class'),
    ('aquatic', 'class'),
    ('predator', 'class'),
    ('toothed', 'class'),
    ('backbone', 'class'),
    ('breathes', 'class'),
    ('venomous', 'class'),
    ('fins', 'class'),
    ('legs', 'class'),
    ('tail', 'class'),
    ('domestic', 'class'),
    ('catsize', 'class')
]

expert_model = BayesianNetwork(expert_edges)

expert_model.fit(train_df, estimator=MaximumLikelihoodEstimator)

print("Экспертная структура создана: все 16 признаков → class")
print("Рёбер:", len(expert_model.edges()))
````

Этот код использует метод поиска с подъемом на холм (Hill Climb Search) для оценки структуры Байесовской сети из данных, используя Байесовский информационный критерий (BIC) в качестве функции оценки.

BIC (Bayesian Information Criterion) — это функция оценки, используемая для сравнения статистических моделей (в данном случае, структур Байесовских сетей).

Цель: Он наказывает модели за излишнюю сложность (большое количество параметров/ребер), одновременно вознаграждая за хорошее соответствие данным.

Принцип: Модели с меньшим значением BIC считаются лучшими.

Формула (концептуально): BIC≈L−k⋅ln(N), где L — логарифмическое правдоподобие модели (насколько хорошо модель объясняет данные), k — количество параметров модели (сложность), а N — количество точек данных.


### 3. Получение вероятностей

Мы хотим получить вероятностные оценки переходов из одного состояние в другое. Для этого необходимо Байесовскую сеть подогнать по данные и получить таблички (CPT).

Conditional Probability Table (CPT) [Таблица Условных Вероятностей (CPT)] — это ключевой элемент, определяющий параметры Байесовской сети.

* Что это: Это таблица, которая содержит вероятность каждого возможного состояния узла (переменной) при условии всех возможных комбинаций состояний его родительских узлов (тех, от которых к нему идут стрелки/ребра).
* Роль: CPTs количественно определяют зависимости и силу связей между переменными, позволяя сети выполнять логический вывод и прогнозирование.
* Простой пример: В CPT для узла "Кашель", родительским узлом которого является "Грипп", будет указана вероятность кашля, если грипп есть (например, 90%), и вероятность кашля, если гриппа нет (например, 5%).

Чтобы получить значения вероятности переходов из разных узлов, нужно оценить эти вероятности. `HillClimbSearch` ищет только структуру, а `MaximumLikelihoodEstimator` и `BayesianEstimator` позволяют оценить вероятность.

````python
# Метод Максимального Правдоподобия
from pgmpy.estimators import MaximumLikelihoodEstimator

model.fit(data, estimator=MaximumLikelihoodEstimator)
````

````python
print("\n1. CPT для узла 'class' в ЭКСПЕРТНОЙ (наивной) модели")
print("   (все признаки — родители класса)")
cpd_class_expert = expert_model.get_cpds('class')
print(cpd_class_expert)

key_nodes = ['milk', 'feathers', 'backbone', 'eggs', 'hair', 'class']
for node in key_nodes:
    cpd = model.get_cpds(node)
    if cpd:
        print(f"\nCPT для узла '{node}':")
        if len(cpd.variables) == 1:
            print("  → Этот узел не имеет родителей (корневой)")
        else:
            print(f"  → Родители: {cpd.variables[:-1]}")
        print(cpd)
        if node == 'class':
            print("  → Обратите внимание: milk=1 → class=1 с вероятностью ≈1.0")
            print("    feathers=1 → class=2 с вероятностью ≈1.0")
````

Вместо того чтобы просто считать частоты (как MLE), Байесовский оценщик вычисляет апостериорное распределение вероятностей (CPT)

````python
from pgmpy.estimators import BayesianEstimator

bayes_model = BayesianNetwork(expert_edges)
bayes_model.fit(
    train_df,
    estimator=BayesianEstimator,
    prior_type='BDeu',          
    equivalent_sample_size=10 
)

print("Готово! Модель переобучена с BayesianEstimator (BDeu, ess=10)\n")

print("CPT для 'class' при BayesianEstimator (сглаженные вероятности):")
print(bayes_model.get_cpds('class'))
````

Просмотр полной таблички

````python
cpt_class = model.get_cpds('class')
print(cpt_class)
````

Просмотр таблиц для конкретных узлов

````python
key_nodes = ['class', 'milk', 'feathers', 'backbone', 'eggs', 'hair']
for node in key_nodes:
    cpd = model.get_cpds(node)
    if cpd is None:
        print(f"Узел '{node}' не найден в модели!")
        continue
    
    print(f"→ CPT для узла: {node}")
    if len(cpd.variables) == 1:
        print("   Это корневой узел (нет родителей)")
    else:
        parents = cpd.variables[:-1]
        print(f"   Родители: {', '.join(parents)}")
    
    print(cpd)
    
    if node == 'class' and 'milk' in cpd.variables:
        print("   Интерпретация: milk=1 → class=1 с вероятностью 1.0 → все млекопитающие имеют молоко")
    if node == 'feathers':
        print("   Интерпретация: если класс = 2 (птицы), то feathers=1 с вероятностью 1.0")
    if node == 'eggs' and 'milk' in cpd.variables:
        print("   Интерпретация: если milk=1 → eggs=0 (млекопитающие не несут яйца)")
    
````

### 4. Визуализация сети через библиотеку `networkx`

````python
import networkx as nx
import matplotlib.pyplot as plt

nx_graph = nx.DiGraph(model.edges())
pos = nx.spring_layout(nx_graph)
plt.figure(figsize=(16, 12))
nx.draw(
    nx_graph,
    pos,
    with_labels=True,
    node_size=4500,
    node_color='lightgreen',
    font_size=12,
    font_weight='bold',
    arrows=True,
    arrowsize=25,
    arrowstyle='->',
    edge_color='darkblue',
    width=2.0
)


key_nodes = ['class', 'milk', 'feathers', 'backbone']
for node in key_nodes:
    if node in nx_graph.nodes:
        nx.draw_networkx_nodes(nx_graph, pos, nodelist=[node], node_color='gold', node_size=6000)

plt.title(
    "Байесовская сеть для классификации животных (Zoo Dataset)\n"
    "Ключевые узлы выделены золотым цветом",
    fontsize=18, fontweight='bold', pad=30
)
plt.axis('off')
plt.tight_layout()
plt.show()
````

### 5. Посмотрим пример инференса

````python
from pgmpy.inference import VariableElimination

infer = VariableElimination(model) 

print("\n 1. У животного есть молоко (milk=1).")
query1 = infer.query(variables=['class'], evidence={'milk': '1'})
print(query1)

print("\n 2. У животного есть перья (feathers=1).")
query2 = infer.query(variables=['class'], evidence={'feathers': '1'})
print(query2)

print("\n 3. Животное НЕ дышит воздухом (breathes=0).")
query3 = infer.query(variables=['class'], evidence={'breathes': '0'})
print(query3)

print("\n 4. Рыбы")
query4 = infer.query(
    variables=['class'],
    evidence={'aquatic': '1', 'fins': '1', 'hair': '0'}
)
print(query4)

print("5. MAP-запрос: какое животное наиболее вероятно при признаках?")
map_query = infer.map_query(variables=['class'], evidence={'milk': '1', 'hair': '1', 'eggs': '0'})
print("Признаки: есть молоко, есть шерсть, не несёт яйца →", map_query)
print("→ Ответ: млекопитающее (class=1) — как и ожидалось!")
````

### 6. Сравнение результатов с baseline-моделью (например, наивный байесовский классификатор из scikit-learn) по метрикам accuracy или log-likelihood.

````python
from pgmpy.estimators import BayesianEstimator
from sklearn.metrics import accuracy_score, log_loss

print("="*85)
print("6. СРАВНЕНИЕ С BASELINE (финальная честная версия)")
print("="*85)

infer_auto = VariableElimination(model)
acc_auto = sum(infer_auto.map_query(['class'], evidence=row.drop('class').to_dict())['class'] == row['class'] 
               for _, row in test_df.iterrows()) / len(test_df)

expert_bayes_model = BayesianNetwork(expert_edges)
expert_bayes_model.fit(train_df, estimator=BayesianEstimator, prior_type='BDeu', equivalent_sample_size=20)

infer_expert_bayes = VariableElimination(expert_bayes_model)
acc_expert_bayes = sum(infer_expert_bayes.map_query(['class'], evidence=row.drop('class').to_dict())['class'] == row['class']
                       for _, row in test_df.iterrows()) / len(test_df)

X_train = train_df.drop('class', axis=1).astype(int)
y_train = train_df['class'].astype(int)
X_test  = test_df.drop('class', axis=1).astype(int)
y_test  = test_df['class'].astype(int)

nb = CategoricalNB(alpha=1.0) 
nb.fit(X_train, y_train)
acc_nb = accuracy_score(y_test, nb.predict(X_test))

print(f"\n{'Модель':<52} {'Accuracy'}")
print("-" * 70)
print(f"{'1. Автоматическая структура (HillClimb + MLE)':<52} {acc_auto:.4f}")
print(f"{'2. Наивная структура + BayesianEstimator (BDeu)':<52} {acc_expert_bayes:.4f}")
print(f"{'3. CategoricalNB (sklearn, Laplace smoothing)':<52} {acc_nb:.4f}")
print("-" * 70)

print(f"ЛУЧШАЯ МОДЕЛЬ: {'BayesianEstimator' if acc_expert_bayes > max(acc_auto, acc_nb) else 'sklearn' if acc_nb > max(acc_auto, acc_expert_bayes) else 'HillClimb'}")
print("\nВЫВОД: при малом объёме данных после удаления дубликатов")
print("       критически важно использовать сглаживание (BayesianEstimator или alpha>0)!")
````
